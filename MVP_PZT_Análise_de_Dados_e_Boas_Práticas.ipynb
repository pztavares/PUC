# Contextualização do Problema

A igualdade de gênero vem sendo um assunto muito discutido ultimamente e é o objetivo de número 5 (ODS 5) dos Objetivos do Desenvolvimento Sustentável (ODS), propostos pela Organização das Nações Unidas (ONU). O ODS 5 propõe-se a “Alcançar a igualdade de gênero e empoderar todas as mulheres e meninas”.

Apesar dos avanços das mulheres na reinvindicação de seus direitos, ainda existem muitas barreiras a serem transpostas. O problema é sentido em todas as áreas profissionais, mas ultimamente tem sido **fortemente citado nas áreas STEM**, que á sigla em inglês usada para *Science, Technology, Engineering, and Mathematics* (em português, Ciências, Tecnologia, Engenharia e Matemática).

O problema vem motivando a sociedade a refletir sobre o assunto:

*   Por que a mulheres são a minoria em STEM?
*   Por que as meninas / mulheres não se interessam pela área de STEM?
*   Por que as mulheres abandonam a área de STEM?
*   Como reter as mulheres em STEM?

Existem diversas teorias que tentam explicar a ausência feminina na área, dentre elas está o **estereótipo de gênero** ("meninos são melhores em matemática e meninas em outras áreas, como leitura"). Segundo essa teoria, tais limitações sociais e culturais poderiam desmotivar as meninas no interesse voluntário por áreas STEM em sociedades menos equalitárias.

Então, Países que estimulam a paridade de gênero teriam maior participação de mulheres nos campos STEM do que os outros Países menos equalitários? Como visualizar um panorama mundial para confirmar essa teoria? Como saber se o País é "atrativo" para as mulheres estudarem os campos STEM?

Para este estudo, a pesquisadora buscou dados sobre educação de nível superior de mulheres nas áreas STEM pelo mundo e realizou cruzamentos de dados com algumas características sociais e financeiras dos Países, a fim de analisar tendências de como esse fenômeno ocorre.

## Levantamento dos dados

A pesquisadora realizou um *merge* entre cinco *datasets* levantados em repositórios especializados para compor o *dataset* final do estudo. São eles:

### **Dataset 1**
Dados do Banco Mundial que apresentam o percentual de graduação em STEM por País, ano e gênero (WORLD BANK, 2025a).

Algumas características importantes:

1.   O datatset possui dados do período de 1998 a 2019. Alguns Países não possuem as informações em todos os anos, como o Brasil (com dados entre 2001 e 2017). Portanto, as informações são apresentadas pelo ano mais recente de cada País;
2.   Nem todos os Países do mundo estão refletidos no dataset: alguns Países não possuem informações sobre mulheres nas STEM (como o Afeganistão), portanto, não existem no dataset.

### **Dataset 2**

Classificações de Países de acordo com o Banco Mundial (WORLD BANK, 2025b):

**1.   Por nível de renda:** de acordo com a renda nacional bruta (RNB) per capita.

  *   *High income* - alta renda
  *   *Upper middle income -* renda média-alta
  *   *Lower middle income -* renda média-baixa
  *   *Low income -* baixa renda

**2.   Por região:**
  *   East Asia & Pacific
  *   Europe & Central Asia
  *   Latin America & Caribbean
  *   Middle East & North Africa
  *   North America
  *   South Asia
  *   Sub-Saharan África

### **Dataset 3**
Dados sobre o Índice de Desenvolvimento Humano (IDH), que avalia o desenvolvimento de um país e não apenas o crescimento econômico (UNDP, 2025b).

*   **IDH Muito Alto** - IDH varia tipicamente de 0,800 a 1,000.
*   **IDH Alto** - IDH varia tipicamente de 0,700 a 0,799.
*   **IDH Médio** - IDH varia tipicamente de 0,550 a 0,699.
*   **IDH Baixo** - IDH varia tipicamente de 0,000 a 0,549.

### **Dataset 4**
Nomes dos países em inglês e português, para maior facilidade de leitura dos resultados por País (SISCOMEX, 2025).

### **Dataset 5**
Dados com o Índice de Desigualdade de Género (IDG), ou *Gender Inequality Index* (GII) em inglês, que é um índice que mede a desigualdade entre homens e mulheres em três dimensões: saúde reprodutiva, empoderamento e mercado de trabalho, segundo o Programa das Nações Unidas para o Desenvolvimento (PNUD). O GII varia de 0 a 1, onde 0 representa igualdade completa e 1 representa a máxima desigualdade (UNDP, 2025a).

## Objetivo do Estudo

Diante do contexto apresentado, este estudo tem como objetivo analisar padrões de um Pais como favorável para as mulheres nas áreas STEM com base em quatro características: Índice de Desenvolvimento Humano (IDH), Índice de Desigualdade de Género (Gender Inequality Index, GII), grupo de renda e região geográfica.

Futuramente, pretende-se construído um modelo para previsão de tendência anual de Países emergentes em STEM para mulheres e permitir que as políticas públicas mundiais possam implementar incentivos que promovam a igualdade de gênero em STEM nas localidades mais carentes.




## Hipóteses do Problema

As hipóteses são as seguintes:

- O Indice de Desenvolvimento Humano (IDH) tem correlação com o Gender Inequality Index (GII), ou seja, um País com IDH alto provavelmente é um país com baixa desigualdade de gênero?

- A desigualdade de gênero é um fenômeno mundial ou concentra-se em alguns grupos de renda ou regiões geográficas específicas?

- Em um País com um Gender Inequality Index (GII) baixo existe maior participação de mulheres nos cursos superiores em STEM?

## Atributos do Dataset

O dataset contém 129 linhas (Países) e 18 atributos:

- ***CodPais*** (sigla do País, de acordo com o Banco Mundial)
- ***PaisIngles*** (nome do País em inglês)
- ***PaisPortugues*** (nome do País em português)
- ***Regiao*** (região do País, de acordo com o Banco Mundial)
- ***GrupoRenda*** (nível de renda do País, de acordo com o Banco Mundial)
- ***AnoSTEM*** (ano de graduação em STEM)
- ***%MulheresSTEM*** (percentual de mulheres graduadas em STEM)
- ***%HomensSTEM*** (percentual de homens graduados em STEM)
- ***IDH*** (IDH do País)
- ***IDHClassificacao*** (classificação do País por seu IDH)
- ***GII*** (Índice de Desigualdade de Género do País)
- ***TaxaMortalidadeMaterna*** (taxa de mortalidade materna do País)
- ***TaxaGravidezAdo*** (taxa de gravidez na adolescência do País)
- ***CadeirasParlamento*** (porcentagem de assentos parlamentares ocupados por mulheres)
- ***MulheresEducacaoSecundaria*** (níveis de escolaridade de ensino médio das mulheres)
- ***HomensEducacaoSecundaria*** (níveis de escolaridade de ensino médio dos homens)
- ***TaxaForcaTrabalhoFem*** (força de trabalho feminina)
- ***TaxaForcaTrabalhoMasc*** (força de trabalho masculina)

# Análise de Dados

Etapa de Análise de Dados Exploratória (EDA), vamos entender a distribuição, as relações e as características das variáveis para as etapas subsequentes de pré-processamento e modelagem.

```python
# Após a identificação do dataset, observa-se que alguns campos foram carregados com o tipo incorreto
# Ajustes nos tipos:
dataset['AnoSTEM'] = dataset['AnoSTEM'].fillna(0).astype(int)
dataset['%MulheresSTEM'] = dataset['%MulheresSTEM'].str.replace(',','.').fillna(0).astype(float)
dataset['%HomensSTEM'] = dataset['%HomensSTEM'].str.replace(',','.').fillna(0).astype(float)
dataset['IDH'] = dataset['IDH'].str.replace(',','.').fillna(0).astype(float)
dataset['GII'] = dataset['GII'].str.replace(',','.').fillna(0).astype(float)
dataset['TaxaMortalidadeMaterna'] = dataset['TaxaMortalidadeMaterna'].fillna(0).astype(int)
dataset['TaxaGravidezAdo'] = dataset['TaxaGravidezAdo'].str.replace(',','.').fillna(0).astype(float)
dataset['CadeirasParlamento'] = dataset['CadeirasParlamento'].str.replace(',','.').fillna(0).astype(float)
dataset['MulheresEducacaoSecundaria'] = dataset['MulheresEducacaoSecundaria'].str.replace(',','.').fillna(0).astype(float)
dataset['HomensEducacaoSecundaria'] = dataset['HomensEducacaoSecundaria'].str.replace(',','.').fillna(0).astype(float)
dataset['TaxaForcaTrabalhoFem'] = dataset['TaxaForcaTrabalhoFem'].str.replace(',','.').fillna(0).astype(float)
dataset['TaxaForcaTrabalhoMasc'] = dataset['TaxaForcaTrabalhoMasc'].str.replace(',','.').fillna(0).astype(float)
```
